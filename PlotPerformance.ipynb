## Load saved module performance from file

```python
import pickle
import matplotlib.pyplot as plt

filename = 'C:/Users/SarithaSiva/AI/AI Final project/all_metrics.sav'
 
print("\n", 50 * "-", "\nLoading metrics from", filename)
metrics = pickle.load(open(filename, 'rb'))
 
print("\n", "metrics after reloading", "\n", 50 * "-", "\n", metrics)
```

```python
#fig, ax = plt.subplots(figsize=(8,5))
fig, ax = plt.subplots(figsize=(10,10))
metrics.plot(kind='barh', ax=ax)
ax.set_title('Bar graph for metrics');
ax.grid();
 
plt.show()
```
